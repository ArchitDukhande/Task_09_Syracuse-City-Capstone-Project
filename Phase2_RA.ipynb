```python
#Cell 1: Setup + Load
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

complaints = pd.read_csv("SPD_Personnel_Complaints_2021_to_Present.csv")

crime_2023_p1 = pd.read_csv("Crime_Data_2023_-_Part_1_Offenses_(With_Lat_%26_Long_Info).csv")
crime_2023_p2 = pd.read_csv("Crime_Data_2023_Part_2_Offenses_With_Lat_and_Long.csv")
crime_2024_p1 = pd.read_csv("Crime_Data_2024_Part_1_Offenses_With_Lat_and_Long_Info.csv")
crime_2024_p2 = pd.read_csv("Crime_Data_2024_Part_2_Offenses_With_Lat_and_Long_Info.csv")
crime_2025_p1 = pd.read_csv("Crime_Data_2025_Part_1_Offenses_With_Lat_and_Long_Info_1184453394549971059.csv")

print("Complaints shape:", complaints.shape)
print("Crime 2023 P1:", crime_2023_p1.shape)
print("Crime 2023 P2:", crime_2023_p2.shape)
print("Crime 2024 P1:", crime_2024_p1.shape)
print("Crime 2024 P2:", crime_2024_p2.shape)
print("Crime 2025 P1:", crime_2025_p1.shape)

```

```python
display(complaints.head())
display(crime_2023_p1.head())
display(crime_2023_p2.head())
display(crime_2024_p1.head())
display(crime_2024_p2.head())
display(crime_2025_p1.head())
```

```python
complaints_missing = complaints.isna().mean().sort_values(ascending=False)
print("Complaints missingness (top 15):")
display(complaints_missing.head(15))
```

```python
crime_missing = crime_2024_p1.isna().mean().sort_values(ascending=False)
print("Crime 2024 Part 1 missingness (top 15):")
display(crime_missing.head(15))
```

```python
complaints["Complaint_Date"] = (
    pd.to_datetime(complaints["Complaint_Date"], errors="coerce", utc=True)
      .dt.tz_convert(None)
)

complaints["Closure_Date"] = (
    pd.to_datetime(complaints["Closure_Date"], errors="coerce", utc=True)
      .dt.tz_convert(None)
)

complaints["year"] = complaints["Complaint_Date"].dt.year
complaints["month"] = complaints["Complaint_Date"].dt.month
complaints["days_to_close"] = (
    complaints["Closure_Date"] - complaints["Complaint_Date"]
).dt.days

print("Parsed date types:")
print(complaints["Complaint_Date"].dtype, complaints["Closure_Date"].dtype)
```

```python
print("Complaints by year:")
display(complaints["year"].value_counts(dropna=False).sort_index())

print("Days to close summary:")
display(complaints["days_to_close"].describe())

display(complaints.describe(include="all"))
```

```python
complaints["year"].value_counts().sort_index().plot(kind="bar")
plt.title("Complaints by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
complaints["month"].value_counts().sort_index().plot(kind="bar")
plt.title("Complaints by Month (All Years)")
plt.xlabel("Month")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
crime_all = pd.concat(
    [crime_2023_p1, crime_2023_p2, crime_2024_p1, crime_2024_p2, crime_2025_p1],
    ignore_index=True
)

print("Combined crime dataset shape:", crime_all.shape)
display(crime_all.head())
```

```python
crime_all["DATEEND"] = pd.to_datetime(crime_all["DATEEND"], errors="coerce")

crime_all["TIMEEND"] = (
    crime_all["TIMEEND"]
    .astype(str)
    .str.replace(r"\.0$", "", regex=True)
    .str.zfill(4)
)

crime_all["crime_datetime"] = pd.to_datetime(
    crime_all["DATEEND"].dt.strftime("%Y-%m-%d") + " " +
    crime_all["TIMEEND"].str[:2] + ":" +
    crime_all["TIMEEND"].str[2:],
    errors="coerce"
)

print("Datetime coverage:", crime_all["crime_datetime"].notna().mean().round(3))
display(crime_all[["DATEEND", "TIMEEND", "crime_datetime"]].head())
```

```python
crime_all["year"] = crime_all["crime_datetime"].dt.year
crime_all["month"] = crime_all["crime_datetime"].dt.month
crime_all["year_month"] = crime_all["crime_datetime"].dt.to_period("M").astype(str)

print("Crimes by year:")
display(crime_all["year"].value_counts(dropna=False).sort_index())
```

```python
missing_pct = crime_all.isna().mean().sort_values(ascending=False)
print("Top missing columns (crime data):")
display(missing_pct.head(15))
```

```python
display(crime_all.describe(include="all"))
```

```python
crime_all["year"].value_counts().sort_index().plot(kind="bar")
plt.title("Crime Incidents by Year (20232025)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
crime_all["month"].value_counts().sort_index().plot(kind="bar")
plt.title("Crime Incidents by Month (All Years Combined)")
plt.xlabel("Month")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
crime_all["year_month"].value_counts().sort_index().plot(kind="line")
plt.title("Crime Incidents Over Time (20232025)")
plt.xlabel("Year-Month")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
crime_all["CODE_DEFINED"].value_counts().head(10).plot(kind="barh")
plt.title("Top 10 Offense Categories (20232025)")
plt.xlabel("Count")
plt.ylabel("Offense Type")
plt.tight_layout()
plt.show()
```

Key Findings (Phase 2)

The complaints dataset contains substantial missingness in several action- and allegation-related fields, which limits the ability to draw reliable case-level conclusions.

Complaint volumes vary across years and months, indicating clear temporal patterns and seasonality in reporting.

The distribution of days to close complaints shows wide variability, suggesting that resolution timelines are uneven across cases.

Crime incidents in the combined 20232025 dataset also exhibit monthly patterns, reinforcing the presence of seasonality at an aggregate level.

A relatively small number of offense categories (CODE_DEFINED) account for a large proportion of recorded crime incidents.
